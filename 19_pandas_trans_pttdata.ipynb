```python
import pandas as pd
import os
```

```python
flist = os.listdir('./ptt_gossiping')
flist
```

```python
for textfile in flist:
    with open('./ptt_gossiping/%s' %(textfile), 'r',encoding ='utf-8') as f:
        data_str = f.read().split('----------split--------------')[1]
        print(data_str)
    print('=============')
```

```python
data =[]
for textfile in flist:
    #if textfile != '.ipynb_checkpoints :'#with縮排進去
    #if textfile == '.ipynb_checkpoints :#不做任何事
        #continue
    with open('./ptt_gossiping/%s' %(textfile), 'r',encoding ='utf-8') as f:
        data_str = f.read().split('----------split--------------')[1]
        data_row = data_str.split('\n')[1:-1] #排除前後空字串[1:-1]
        data.append(data_row)#把每一列一直加進去，所以每一篇文章就一個list，一直新增上去
       #print(data_str)
#         print(data_row)
#     print('=============')
data
```

```python
columns = ['推','噓','分數','作者','標題','時間'] #加上行頭
df = pd.DataFrame(data=data,columns=columns) #變成dataframe的型態
df
```

```python
#製作一個過濾冒號的過濾器
def tmpfilter(s):
    result = s.split(': ')[1]
    return result
```

```python
tmpfilter('推: 13')
```

```python
#備份 用 .copy()函數
#df_copied = df   這樣子寫 原稿跟複製搞都會隨之變動

df_copied = df.copy()
```

```python
#只找推的欄位
df_copied['推']
```

```python
#把推的欄位放進過濾器.apply()

df_copied['推'] = df_copied['推'].apply(tmpfilter)
```

```python
df_copied
```

```python
df_copied = df.copy()
df_copied
```

```python
columns
```

```python
#把每一列都放入過濾器李
for c in columns:
    df_copied[c] = df_copied[c].apply(tmpfilter)
df_copied
```

```python
df_copied2 = df.copy()
df_copied2
```

```python
#用lambda的形式
for c in columns:
    df_copied2[c] = df_copied2[c].apply(lambda s: s.split(': ')[1])
df_copied2
```

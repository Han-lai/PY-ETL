```python
import jieba
import os

jieba.load_userdict('./mydict.txt')
```

```python
flist = os.listdir('./pttGossip')
flist
```

```python
sb = ''
for textfile in flist:
    if textfile == '.ipynb_checkpoints':
        continue
    with open('./pttGossip/%s'%(textfile), 'r', encoding='utf-8') as f:
        sb += f.read().split('---split---')[0]
    
sb # 結果是一個字串
```

```python
s_cut = jieba.cut(sb) #對sb進行斷詞
```

```python
for w in s_cut:
    print(w)
```

```python
for w in s_cut:
    print(w)
#為空值，裡面的東西取過一次就沒了，是用yield寫出來的
```

```python
s_cut = jieba.cut(sb)
wc = {} #宣告一個空的字典

for w in s_cut:
    if len(w) < 2:
        continue
    if not w in wc:
        wc[w] = 1
    else:
        wc[w] += 1 #如果字典裡沒有這個詞就回傳1,若有這個詞就+1

wc
```

```python
wc_list = [(w, wc[w]) for w in wc] #雙層迴圈
wc_list
```

```python
wc_list = []
for w in wc:
    tmpdata = tuple([w, wc[w]])
    wc_list.append(tmpdata)
wc_list
```

```python
wc_list.sort(key=lambda t: t[1]) #由小排到大
wc_list
```

```python
wc_list.sort(key=lambda t: t[1], reverse=True) #由大排到小
wc_list
```
